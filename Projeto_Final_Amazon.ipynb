```python
# Importando a ferramenta de verificação de qualidade de dados
import pandera as pa
```

```python
# verificar os tipos de dados
df.dtypes
```

```python
schema = pa.DataFrameSchema({'nome_do_filme': pa.Column(pa.String),
                             'linguagem': pa.Column(pa.String),
                             'classificaçao_do_publico': pa.Column(pa.String),
                             'tempo_de_execução': pa.Column(pa.String),
                             'ano_de_lançamento': pa.Column(pa.String),
                             'classificaçao_etária': pa.Column(pa.String),
                             'enredo': pa.Column(pa.String),
                            })
```

```python
# Validar o DataFrame
schema.validate(df)
```
